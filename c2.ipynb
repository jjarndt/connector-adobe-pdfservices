**Setup**

```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request

DATASET_DIR = Path("datasets")
TARBALL_PATH = DATASET_DIR / "housing.tgz"
HOUSING_CSV_PATH = DATASET_DIR / "housing/housing.csv"
HOUSING_URL = "https://github.com/ageron/data/raw/main/housing.tgz"

def download_housing_data():
    if not TARBALL_PATH.is_file():
        DATASET_DIR.mkdir(parents=True, exist_ok=True)
        urllib.request.urlretrieve(HOUSING_URL, TARBALL_PATH)
        print("Data loaded...")
        with tarfile.open(TARBALL_PATH) as housing_tarball:
            print("Extract data...")
            housing_tarball.extractall(path=DATASET_DIR)

def load_housing_data():
    if not HOUSING_CSV_PATH.is_file():
        download_housing_data()
    return pd.read_csv(HOUSING_CSV_PATH)

housing = load_housing_data()
```
